**Madidas de posição e de dispersão**


```python
import numpy as np 
from scipy import stats

# Atribuição de valores
V = [15.425,21.385,11.105,19.554,21.894,7.527,21.790,19.554,20.555,13.614,8.260,23.938]
```

```python
#Medidas de posição

print("Média amostral =",np.mean(V))
print('Mediana =',np.median(V))
print('Moda:', stats.mode(V))
print("Quartis:",np.quantile(V,[0,0.25,0.5,0.75,1]))
print('Percentil de 14%:',np.quantile(V,0.14))
```

```python
#Medidas de disperção

print('Mínimo=',np.min(V),';Máximo=',np.max(V))
print("Desvio padrão POPULACIONAL=",np.std(V))
print("Desvio padrão AMOSTRAL=",np.std(V,ddof=1))
CV = 100*(np.std(V,ddof=1)/np.mean(V))
print("Coeficiente de variação", CV, "%") 
```
