```
# Tabelas de contagem das variáveis quantitativas (ao menos 3)
print(Genero.value_counts())
print("")
print(Municipio.value_counts())
print("")
```

```
# Tabela contingência (tabulação cruzada): gênero e município
pd.crosstab(Genero,Municipio)
```

```
# Tabela de contingência com os totais gerais (OPCIONAL)
pd.crosstab(Genero,Municipio,margins = True)
```

```
# Tabela de contingência com as frequências relativas ao total geral (OPCIONAL)
pd.crosstab(Genero,Municipio,normalize= True)
```

```
# Gráficos de barras das tabelas de contagem usando o
# método 'plot' da biblioteca Matplotlib. 
# Observe o uso do encadeamento de métodos!
Genero.value_counts().plot(kind = 'bar')
plt.title("Contagem de gêneros")
plt.show()
```
