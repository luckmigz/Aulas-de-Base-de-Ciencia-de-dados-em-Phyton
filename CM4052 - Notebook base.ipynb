```python
#import 
import pandas as pd
import matplotlib.pyplot as plt

#declare
#o disco deve ser'montado' e o caminho trocado
df = pd.read_csv('/content/drive/MyDrive/Aula Bases Mat p  Ciencias de dados/DataSet/pesquisa.csv')
```

```python
# Exemplo de acesso a uma coluna de um DataFrame
C1 = df['Genero']
print(type(C1))
print(C1)
```

```python
# Exemplos de acesso a uma lista de colunas de um DataFrame
C2 = df[['Idade','Religiao']]
print(type(C2))
print(C2)
```

```python
# Exemplos de cálculos estatísticos
MI = df['Idade'].mean()
MP = df['Peso'].mean()
MA = df['Altura'].mean()
print("Média das idades =",MI)
print("Média dos pesos =",MP)
print("Média das alturas =",MA)

# Coeficiente de variação (%) das idades
CVI = 100*df['Idade'].std()/df['Idade'].mean()
print("Coeficiente de variação(%) das idades = ", CVI)
```

```python
# Histograma das alturas
df.hist(column='Altura',bins = 5)
plt.title("Alturas (em cm)")
plt.ylabel("Contagens")
plt.show()

# Histograma das Idades
df.hist(column = 'Idade', bins = 6, grid = False)
plt.xlabel("Anos")
plt.title('Idades')
plt.show()

# Boxplot das idades
df.boxplot(column = 'Idade')
plt.title("Idades")
plt.ylabel("Anos")
plt.show()
```

```python
# Boxplots das idades, separados por gênero
df.boxplot(column = 'Idade',by = 'Genero',grid = False)
plt.ylabel("Anos")
plt.show()
```

```python
# Diagrama de dispersão (um dos possíveis)
df.plot.scatter(x='Idade', y='Altura')
plt.title("Alturas x Idades")
plt.xlabel("Anos")
plt.ylabel("cm")
plt.show()

df.plot.scatter(x='Peso',y='Altura')
plt.title("Peso x Altura")
plt.xlabel("Kg")
plt.ylabel("cm")
plt.show()
```

```python
# Tabelas de contagem
print(df['Genero'].value_counts())
print(df['Esporte'].value_counts())
print(df['Religiao'].value_counts())
print(df['Sono'].value_counts())
print(df['Municipio'].value_counts())
```

```python
#Tabela de contigência 
pd.crosstab(df['Genero'],df['Municipio'])

#Gráfico de barra
pd.crosstab(df['Genero'],df['Municipio']).plot(kind = 'bar')
```
